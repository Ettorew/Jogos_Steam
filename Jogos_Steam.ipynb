```python
import pandas as pd
df = pd.read_csv("/content/2023 - Year Stats - SteamSpy - All the data and stats about Steam games.csv", delimiter=';')
df
```

```python
df_limpo = df[df['Playtime (Median)'] != '00:00 (00:00)']
df_limpo
```

```python
df_limpo = df[df['Score rank(Userscore / Metascore)'] != 'N/A (N/A)']
df_limpo
```

```python
df_limpo = df.drop(columns=['Score rank(Userscore / Metascore)', 'Playtime (Median)'])
df_limpo
```

```python
df_limpo['Owners'].unique()
```

```python
df_limpo[['Min_Owners(m)', 'Max_Owners(m)']] = df_limpo['Owners'].str.replace('\xa0', ' ').str.split(' .. ', expand=True)

# Removendo as vírgulas e convertendo para números
df_limpo['Min_Owners(m)'] = df_limpo['Min_Owners(m)'].str.replace(',', '').astype(int)/1000
df_limpo['Max_Owners(m)'] = df_limpo['Max_Owners(m)'].str.replace(',', '').astype(int)/1000
df_limpo = df_limpo.drop(columns='Owners')
df_limpo
```

```python
df_limpo.isna().sum()
```

```python
import re

df_limpo['Release date'] = df_limpo['Release date'].apply(lambda x: re.sub(r',.*', '', x))
df_limpo
```

```python
dic = {
    'Jan' : 1,
    'Feb' : 2,
    'Mar': 3,
    'Apr' : 4,
    'May' : 5,
    'Jun' : 6,
    'Jul' : 7,
    'Aug' : 8,
    'Sep' : 9,
    'Oct' : 10,
    'Nov' : 11,
    'Dec' : 12
}

def extract_month(date_str):
    # Extrair as primeiras 3 letras da string (abreviação do mês)
    month_abbr = date_str.split()[0]
    return dic.get(month_abbr, None)  # Retorna o número do mês, ou None se não encontrar

# Aplicar a função para criar a nova coluna 'month'
df_limpo['Month'] = df_limpo['Release date'].apply(extract_month)

# Exibir o DataFrame final
df_limpo
```

```python
df_limpo = df_limpo.dropna()
df_limpo = df_limpo.drop(columns='#')

def convert_price(price_str):
    if(price_str == 'Free'):
      return 0.0
    # Remover o símbolo de espaço e o símbolo de moeda '$', depois substituir a vírgula por ponto
    cleaned_price = price_str.replace(' $', '').replace(',', '.')
    return float(cleaned_price)

# Aplicar a função à coluna 'Price'
df_limpo['Price'] = df_limpo['Price'].apply(convert_price)
df_limpo
```

```python
print('Média preço dos jogos lançados em 2023: $%.2f' % df_limpo['Price'].mean())
print('Mediana preço dos jogos lançados em 2023: $%.2f' % df_limpo['Price'].median())
print('Desvio Padrão do preço dos jogos lançados em 2023: %.2f' % df_limpo['Price'].std())
print('Primeiro quartil da estimativa máxima de possuidores dos jogos lançados em 2023: %.0f' % df_limpo['Max_Owners(m)'].quantile(0.25))
print('Mediana da estimativa máxima (em milhares) de possuidores dos jogos lançados em 2023 %d' %df_limpo['Max_Owners(m)'].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a média de Price para cada valor único de Min_Owners(m)
media_price = df_limpo.groupby('Min_Owners(m)')['Price'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(data=media_price, x='Min_Owners(m)', y='Price', palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('Média do Preço por Estimativa Mínima de Possuidores dos Jogos', fontsize=14)
plt.xlabel('Estimativa Mínima de Possuidores (milhares)', fontsize=12)
plt.ylabel('Média do Preço (em $)', fontsize=12)

# Exibir o gráfico
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização, se necessário
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular a média de Price para cada valor único de Min_Owners(m)
media_price = df_limpo.groupby('Min_Owners(m)')['Price'].median().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10,6))
sns.barplot(data=media_price, x='Min_Owners(m)', y='Price', palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('Mediana do Preço por Estimativa Mínima de Possuidores dos Jogos', fontsize=14)
plt.xlabel('Estimativa Mínima de Possuidores (milhares)', fontsize=12)
plt.ylabel('Mediana do Preço (em $)', fontsize=12)

# Exibir o gráfico
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização, se necessário
plt.tight_layout()
plt.show()
```

```python
contagem_owners = df_limpo['Min_Owners(m)'].value_counts()

# Criar o gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(contagem_owners, labels=contagem_owners.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Adicionar título
plt.title('Distribuição das Observações para Cada Estimativa Mínima de Possuidores', fontsize=14)

# Exibir o gráfico
plt.axis('equal')  # Garante que o gráfico seja um círculo perfeito
plt.show()
```

```python
# Calcular a receita total (Preço * Min_Owners(m)) para cada linha
df_limpo['Total_Revenue'] = df_limpo['Price'] * df_limpo['Min_Owners(m)']

# Ordenar os jogos pela receita total em ordem decrescente
top_10_games = df_limpo[['Game', 'Total_Revenue']].sort_values(by='Total_Revenue', ascending=False).head(10)

# Criar o gráfico de barras
plt.figure(figsize=(12,6))
plt.barh(top_10_games['Game'], top_10_games['Total_Revenue'], color='skyblue')

# Adicionar título e rótulos aos eixos
plt.title('Top 10 Jogos com Maiores Receitas', fontsize=14)
plt.xlabel('Receita Total Mínima (em $)', fontsize=12)
plt.ylabel('Jogo', fontsize=12)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Agrupar pelos meses e calcular a soma de Min_Owners(m) para cada mês
revenue_by_month = df_limpo.groupby('Month').agg({'Min_Owners(m)': 'sum', 'Game': 'count'}).reset_index()

# O total de jogos lançados será o número de jogos (contagem de 'Game')
# O total de 'Min_Owners(m)' será a soma da estimativa mínima de possuidores
revenue_by_month.columns = ['Month', 'Total_Min_Owners(m)', 'Games_Launched']

# Criar o gráfico de barras
plt.figure(figsize=(12,6))
plt.bar(revenue_by_month['Month'], revenue_by_month['Total_Min_Owners(m)'], color='skyblue')

# Adicionar título e rótulos aos eixos
plt.title('Número Total de Jogos comprados por Mês', fontsize=14)
plt.xlabel('Mês de Lançamento', fontsize=12)
plt.ylabel('Número Mínimo Total de Jogos comprados', fontsize=12)

# Adicionar a legenda mostrando o número de jogos lançados no mês
for i, row in revenue_by_month.iterrows():
    plt.text(row['Month'], row['Total_Min_Owners(m)'], f'{row["Games_Launched"]} Jogos',
             ha='center', va='bottom', fontsize=10)

# Exibir o gráfico
plt.tight_layout()
plt.xticks(rotation=45)  # Ajuste para que os meses apareçam de forma legível
plt.show()
```

Em cima é o número de jogos lançados no mês, eixo x o mês, eixo Y as vendas totais dos jogos lançados no mês em questão
